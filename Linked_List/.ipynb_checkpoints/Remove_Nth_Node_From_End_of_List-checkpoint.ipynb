Given the head of a linked list, remove the nth node from the end of the list and return its head.

## Example 1:



__Input__: head = [1,2,3,4,5], n = 2  
__Output__: [1,2,3,5]


## Example 2:
__Input__ : head = [1], n = 1  
__Output__: []


## Example 3:

__Input__: head = [1,2], n = 1  
__Output__: [1]

```python
def removeNthFromEnd(self, head, n):
    size = 1
    cur = p = head
    while cur.next:
        size += 1
        cur = cur.next
        if size > n + 1:
            p = p.next
    if size == n:
        return head.next
    else:
        p.next = p.next.next
        return head
```
